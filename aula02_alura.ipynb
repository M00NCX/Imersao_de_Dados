```python
## DESAFIO 03 + 07

tipos_de_tratamento = dados.tratamento.value_counts()
labels = (dados.tratamento.unique())
explode = (0, 0.1)
cores = ['salmon', 'springgreen']


plt.title('Tipos de Tratamentos', fontsize=16, fontweight='bold')
plt.pie(tipos_de_tratamento, labels=labels, explode=explode, autopct='%2.2f%%', colors=cores)
plt.tight_layout()
plt.axis('equal')
plt.show()
```

```python
## DESAFIO 07
dados['tempo'].value_counts().plot.bar()
tempo_hrs = dados.tempo.value_counts()

plt.title('Tempo de exposição', fontsize=16, fontweight='bold')
plt.bar(tempo_hrs, height= 0.6, width=0.8)

tempo_hrs.plot.bar(color=['springgreen','mediumseagreen','forestgreen'])
plt.xlabel('Tempo em horas', fontsize=12)
plt.ylabel('Quantidade de tratamentos', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
bxg = sns.boxplot(y='g-0' , x='tratamento', data=dados, palette='Blues')
bxg.set_xlabel('Tratamento')
bxg.set_title('Boxplot G-0 x Tratamento', fontsize=14, fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(8,6))
bxc = sns.boxplot(y='c-0' , x='tratamento', data=dados, palette='Purples')
bxc.set_xlabel('Tratamento')
bxc.set_title('Boxplot C-0 x Tratamento', fontsize=14, fontweight='bold')
plt.show()
```

```python
fig, (dg, dc) = plt.subplots(1, 2, figsize=(10,8))

sns.boxplot(y='g-0' , x='dose', data=dados, palette = 'crest', ax=dg)
dg.set_xlabel('Dose')
dg.set_title('Boxplot G-0 x Dose', fontsize=14, fontweight='bold')

sns.boxplot(y='c-0' , x='dose', data=dados, palette = 'viridis', ax=dc)
dc.set_xlabel('Dose')
dc.set_title('Boxplot C-0 x Dose', fontsize=14, fontweight='bold')

plt.show()
```

```python

dt = sns.boxplot(y='g-0' , x='tempo', data=dados, palette='gray')
dt.set_xlabel('Tempo(h)')
dt.set_ylabel('G-0')
dt.set_title('G0 x Tempo', fontsize=14, fontweight='bold')
```
